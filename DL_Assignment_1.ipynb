```python
from tensorflow.keras.datasets import fashion_mnist

(X_train, y_train), (X_test, y_test) = fashion_mnist.load_data()
```

```python
X_train = X_train/255.0 # we divide it so that we can get the pixel values from range of 0 - 1
X_test = X_test/255.0 # 0 - black and 1 - white
```

```python
X_train = X_train.reshape(-1, 28*28)
X_test = X_test.reshape(-1, 28*28)   #resizing the data from 2d matrics to 1d in pixel
# for mapping the pixel -1 all samples reshape 784 1D array
```

```python
import tensorflow as tf
model = tf.keras.models.Sequential()
# from input layer to hidden layer in sequence
# dense is the hidden lay totally 3 layer iniside and units are number of nurons presents
#goes on decreasing to 10 as there are only 10 classes at the end
# Activation functions Relu - Binary classsification and Softmax - multiclass classification
model.add(tf.keras.layers.Dense(units=256, activation='relu', input_shape=(784,)))
model.add(tf.keras.layers.Dropout(0.25))
# droupout avoiding overfitting so dropout required
model.add(tf.keras.layers.Dense(units=128, activation='relu')) # binary - relu
model.add(tf.keras.layers.Dropout(0.25))
model.add(tf.keras.layers.Dense(units=10, activation='softmax')) #multiclass for softmax
```

```python
model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['sparse_categorical_accuracy'])
model.fit(X_train, y_train, epochs=20)
# Adam optimiser is used
# Adam (Adaptive Moment Estimation)
# optimizer combines the advantages of Momentum and RMSprop techniques to adjust learning rates
# during training. It works well with large datasets and complex models
# because it uses memory efficiently and adapts the learning rate for each parameter automatically.
```

```python
import numpy as np
y_test = np.array(y_test, dtype=int)
test_loss, test_accuracy = model.evaluate(X_test, y_test)
```

```python
import matplotlib.pyplot as plt

class_names = ['T-shirt/top', 'Trouser', 'Pullover', 'Dress', 'Coat',
               'Sandal', 'Shirt', 'Sneaker', 'Bag', 'Ankle boot']

y_pred = model.predict(X_test)

plt.figure(figsize=(10, 10))
for i in range(25):
    plt.subplot(5, 5, i + 1)
    plt.xticks([])
    plt.yticks([])
    plt.grid(False)
    plt.imshow(X_test[i].reshape(28, 28), cmap=plt.cm.binary)
    predicted_label = np.argmax(y_pred[i])
    true_label = y_test[i]
    if predicted_label == true_label:
        color = 'blue'
    else:
        color = 'red'
    plt.xlabel(f"{class_names[predicted_label]} ({class_names[true_label]})", color=color)
plt.show()
```
